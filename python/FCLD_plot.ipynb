```python
import json

def make_data_for_line(left, right, prob, y_axis_data, depth = 1):
  output = []
  with open('/content/drive/MyDrive/FCLD_data/output_95_{}.json'.format(depth)) as f:
    j_list = json.load(f)

    for i in j_list:
      if(i['input']['left'] == left and i['input']['right'] == right ):
        output.append(i)
  x = []
  y = []
  for o in output:
    if(o['input']['prob'] == prob):
      x.append(o['input']['param'])
      y.append(o['output'][y_axis_data])
  return x, y
```

```python
import matplotlib.pyplot as plt
import numpy as np


# make data
x1, y1 = make_data_for_line(15, 20, 70, 'edge', depth = 1)
x2, y2 = make_data_for_line(15, 20, 70, 'edge', depth = 2)


# plot
fig, ax = plt.subplots()

ax.plot(x1, y1, linewidth=2.0,  marker='o', color = 'red')
ax.plot(x2, y2, linewidth=2.0,  marker='o', color = 'blue')


plt.xlabel('param')
plt.ylabel('corss value')

plt.show()
```
